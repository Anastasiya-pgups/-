4. Задание (в программе):
 Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10, 10, 0.001)
```

```python
y1 = np.cos(1 * x)
y2 = np.cos(1.2 * x)
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.plot(x, y1, label = 'cos(1*x)', color='green')
plt.plot(x, y2, label = 'cos(1.2*x)', color='red')
plt.title('График f(t)')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.grid(True)
plt.axis([-10, 10, -1.2, 1.2]) 
plt.legend(loc='lower right')
plt.show()
```
